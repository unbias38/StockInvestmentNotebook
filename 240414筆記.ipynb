中東衝突再起，伊朗攻擊以色列。  
  
• 總結：假若美國對伊朗實施更為嚴格的經濟制裁措施，則預估伊朗原油產量將因此減少70-125萬桶/日，並導致2Q24布蘭特原油均價上漲至每桶93美元，QoQ+15%，2024年每桶90美元，YoY+10%。  
  
主要原因如下：  
  
1. 最新3月伊朗原油產量為325萬桶/日。根據2018-2020年期間，美國政府對伊朗實施制裁，並導致原油產量降到200-255萬桶/日為計算基礎，則預估美國一旦對伊朗加大制裁力道，伊朗原油產量將因此減少70-125萬桶/日（在伊朗未封鎖霍爾木茲海峽下）約占全球原油產量0.7-1.3%。  
2. OPEC+等國1-3月減產執行率超過60%，原油供應得到較好的限制。  
3. 夏天出行需求增加，預估2Q24全球原油需求季增5%。  
  
• 根據上述因素，預估全球原油庫存缺口在2Q24將來到150-200萬桶/日。按原油供需差1%，油價波動10%來計算，則在伊朗面臨制裁的情況下，預估2Q24布蘭特原油均價將來到US\\$93/b，QoQ+15%（原預估US\\$85/b）2024年US\\$90/b，YoY+10%（原預估US\\$84/b）  
  
• 對下游塑化產業而言，雖然進入2Q24是傳統需求旺季，同時2Q24也是亞洲裂解廠歲修旺季（煉油供應減少849萬噸，乙烯供應減少118萬噸）預估導致亞洲下游衍生物（PE/PP/EG）供應因此減少1-4%，市場供需得以轉好。惟在原油價格上漲（季增15%）也將進一步墊高塑化業者生產成本下，壓縮塑化產品利差，不利亞洲輕油裂解獲利表現，維持臺灣一線/二線塑化業者台塑（1301）南亞（1303）台化（1326）亞聚（1308）中立投資想法不變。
