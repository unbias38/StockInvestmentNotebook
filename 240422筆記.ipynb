### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤後快訊】短線急跌，中期支撐  
上週台股在FED降息可能推遲、台積電對半導體展望下修、地緣政治風險，多個利空夾擊之下，出現連續性的下跌修正，累計跌幅超過千點，也符合前兩周的規劃，有興趣可以自行複習前兩周的解盤和影音教學，就不過多贅述。  
接下來投資人關心的不外乎3個問題，首先是指數短線是不是已經止跌？由於國際股市和加權指數由於急跌，部分完成頭部跌幅滿足，而多數都已經中期支撐，所以短線有機會止跌。  
其次是會不會反彈，以及反彈反彈的幅度？短線上可以先規劃出兩個壓力區間，藍色為弱勢反彈壓力，紅色為強勢反彈壓力。最後是中長期續跌空間還有多少？這個部分仍須持續觀察，因為以跌幅滿足觀察已經到黃金分割率0.382-0.5，但頭部尚未到滿足區。  
接下來技術面的中期觀察則是季線支撐和扣抵值的變化，長期觀察則是月KD指標是否出現死亡交叉，且脫離鈍化區。並且需要同步觀察外資期貨未平倉空單口數和台幣升貶。  
指數尚未有效站回紅色壓力之前，持股比重高者，反彈宜逢高減碼，心理壓力會比較低。倘若高檔有積極減碼，甚至已經空手，可以用1-3成資金參與，由於暫時判斷只是反彈不是翻多，而且支撐還需要測試，所以指數短線會進入小區間震盪，多空操作個股為主。 (出處：K線捕手楊忠憲粉絲團)
