### 【[K線捕手](https://www.facebook.com/kevinyang20180801)盤前快訊】多方攻擊，留意要件  
加權指數昨天僅小漲34點，單一K棒收十字線，代表多空拉鋸行情膠著，櫃買指數也是類似的現象。籌碼面，法人買賣不大，融資持續攀升，主力指標偏弱，持續呈現觀望態勢。指數盤整加上量能只有2914億，台股動能不足，多數個股也都陷入橫盤或是小回，不如前2周好操作。  
但昨天伴隨歐美股市偏多發展，夜盤期指上漲92點，微幅越過盤整壓力區，有利於今天加權指數續攻，也符合周末以來的看法，由於20日線扣抵，「本周是多方表現最好的時間點」。  
只是畢竟現在位階偏高，加上先前又曾出現假突破和背離現象，既然要過高，就要給予更嚴格的條件，否則又容易再度曇花一現。  
突破3要件是價格、量能、時間，過高累計漲幅至少要200點以上，成交量要放大到3500億以上，能站穩17516前高3天以上。這樣才能達到多方續攻的條件，而且之後即使回測也要守穩17516，不然又容易有第3次背離風險。  
可以留意基期低的多方類股族群有造紙、橡膠、鋼鐵、汽車、航運。台股指數上有鍋蓋、下有鐵板，尚未分出趨勢，不過只要悶鍋的蓋子一旦打開，或是鐵板一旦跌破，多空趨勢會走得相當快，靜待趨勢明朗再決定積極操作的方向。 (出處：K線捕手楊忠憲粉絲團)
