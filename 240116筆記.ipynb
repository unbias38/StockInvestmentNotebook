[我們都希望任何時候都能「多賺少賠」](https://www.facebook.com/kevinyang20180801/posts/pfbid0kwhzsJXMJYpgQyaq7oiEU4d6hNhhQjApmhwN8P5PG6qo2xNJyezr1oxvNyqzXi4dl)，可惜很多時候外在條件注定只能「多賺多賠」，或「少賺少賠」，既然你想避免「多賠」，就一定要捨棄「少賺」的機會。比如在空頭的逆勢操作，正常來說就是減少操作，雖然「少賺」，但可以避免「多賠」。但有人就是會心存僥倖，99檔股票都在跌，我一定可以找到那檔唯一會漲的，當然有時候可以，只是這一次僥倖賺到了，那下次呢？再下次呢？再下下次呢？如果每次都心存僥倖，又沒控制好資金比重跟風險，有一天會吃大虧的。
