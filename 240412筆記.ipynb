紀律是有步驟的行為，從看到交易訊號，然後接著要看的懂這些訊號代表的意義，再來是必須相信這些訊號，緊接著是擬定策略，最後才是紀律執行，但是倘若沒有前置準備，就根本沒有紀不紀律的問題，只是用聽消息、感覺，甚至幻覺買賣股票而已。
