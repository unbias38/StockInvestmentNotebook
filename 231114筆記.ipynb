```python
##### 20211221(20231114) #####

# PLUSExcel
df = pd.read_excel('./data/SINO_2465_231114.xlsx')

# K()
df_k_line = df.drop(columns=['SMA5','SMA10','SMA20','SMA60','MA5','MA10'])

# 
df_sma = df.drop(columns=['','','','','','MA5','MA10']) 

# KColumns
df_k_line = df_k_line.rename(columns={'':'Date','':'Open','':'High','':'Low','':'Close','':'Volume'})

# Columns
df_sma = df_sma.rename(columns={'':'Date'})

# KDateIndex
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# DateIndex
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
def date_to_index(df_in,date_in) :
    number_array = df_in.index == date_in
    idx = 0
    for number_index in number_array :
        if number_index == True :
            break
        idx = idx + 1
    return idx
```

```python
# 
neckline_x_date = ['2022-11-08','2023-10-16']
neckline_x = [date_to_index(df_k_line,neckline_x_date[0]),date_to_index(df_k_line,neckline_x_date[1])]
neckline_y = [df_k_line.loc[neckline_x_date[0]]['Open'],df_k_line.loc[neckline_x_date[1]]['Close']]
slope,intercept = np.polyfit(neckline_x,neckline_y,1)
neckline_start_date = '2022-06-07'
neckline_start_price = slope * date_to_index(df_k_line,neckline_start_date) + intercept
neckline_end_date = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
neckline_end_price = slope * date_to_index(df_k_line,neckline_end_date) + intercept
print('{}  {:.2f}  {}  {:.2f}'.format(neckline_start_date,neckline_start_price,neckline_end_date,neckline_end_price))

# 
bottom_price_date = '2022-09-30'
breakthrough_date = '2023-10-25'
bottom_price = df_k_line.loc[bottom_price_date]['Open']
on_neckline_price = slope * date_to_index(df_k_line,bottom_price_date) + intercept
breakthrough_price = slope * date_to_index(df_k_line,breakthrough_date) + intercept
print('{}  {:.2f}'.format(breakthrough_date,breakthrough_price))
target_price = (on_neckline_price - bottom_price) + breakthrough_price
print('  {:.2f}'.format(target_price))

# K
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_start_price),(neckline_end_date,neckline_end_price)],
    [(bottom_price_date,bottom_price),(bottom_price_date,on_neckline_price)],
    [(breakthrough_date,breakthrough_price),(breakthrough_date,target_price)]
                     ]

# 
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.4,color='xkcd:royal purple'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=0.4,color='xkcd:violet')
            }

# K
kwargs = dict(type='candle', style=s, figratio=(19,10), addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=(1.0,10,10), colors='xkcd:orange yellow', alpha=0.6))

mpf.plot(df_k_line,**kwargs)
```
