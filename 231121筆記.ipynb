可能有投資人會認為，既然現在已經趨勢翻多，為什麼不積極重壓，仍然只使用3-5成資金？首先是趨勢的週期性，現在短期確實是強勢偏多，但中期還在高檔箱型，換言之就是盤整，自然應當多觀望，何況並不是沒持有做多的部位，只是資金比較有所控制，而且也沒有做空。  
其次是選股難度，事實上每天從100檔強勢股中，在不重複的前提之下，最多只能新增2-3檔觀察個股，這意味著選股難度極高，多方的紮實度仍有疑慮。我想大家在選股時，應該有明顯感覺的到。  
最後是位階的問題，倘若今天是去年2022年第四季，從18619高檔一路修正12629，低位階又剛剛翻多，當然資金比重可以積極放大，但現在已經從12629一路上漲到17000點上下，位階相對值就偏高，位階高就是風險。  
所以解盤不是只看漲跌幅，是看綜合趨勢、位階、型態，因此還是維持原本的策略，3-5成資金做多，持股檔數也是3-5檔，這樣一旦走勢遇壓轉弱，要調節會比較快，即使真的很幸運的持續攻高，中期趨勢正式翻多，想要立即增持部位，靈活度也較高，整體操作可攻可守。  

分享一下自己目前的操作狀況，根據周末的規劃，採漸進式減碼，已經賣出的部位，先維持現金，所以作多的個股只剩下兩檔，雖然指數在中期的多空交界拉鋸，尚未分出勝負，但心情算是蠻輕鬆的。  
維持原本的看法，未來真的有效突破壓力，再評估要不要加碼資金和檔數。反過來，如果突破失敗、或是雖然過卻有高檔背離現象，就盯好剩下兩檔個股的支撐，跌破就離場，等待修正完畢。

```python
##### 使用2019年11月5日到今日收盤(2023年11月21日)之日線圖資料 #####

# 載入從「好神通PLUS」輸出的Excel檔
df = pd.read_excel('./data/SINO_1515_231121.xlsx')

# 保存K棒的基本資訊(開、高、低、收、量)
df_k_line = df.drop(columns=['SMA5','SMA20','SMA120','MA5','MA10'])

# 保存價的均線
df_sma = df.drop(columns=['開盤價','最高價','最低價','收盤價','成交量','MA5','MA10']) 

# 將K棒的Columns的名稱由中文改為英文
df_k_line = df_k_line.rename(columns={'時間':'Date','開盤價':'Open','最高價':'High','最低價':'Low','收盤價':'Close','成交量':'Volume'})

# 將價的均線的Columns的名稱由中文改為英文
df_sma = df_sma.rename(columns={'時間':'Date'})

# 將K棒的Date設為Index
df_k_line.set_index(df_k_line['Date'],inplace=True)
df_k_line = df_k_line.drop(columns=['Date'])

# 將價的均線的Date設為Index
df_sma.set_index(df_sma['Date'],inplace=True)
df_sma = df_sma.drop(columns=['Date'])
```

```python
# 設定頸線
neckline_start_date = '2022-01-26'
neckline_end_date = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
neckline_price = df_k_line.loc['2023-09-15']['Open']

# 設定壓力
resistance_start_date = '2021-04-09'
resistance_end_date = df_k_line.iloc[-1].name.strftime("%Y-%m-%d")
resistance_price = df_k_line.loc['2021-05-17']['Close']
```

```python
# 設定K線格式
mc = mpf.make_marketcolors(up='xkcd:light red', down='xkcd:almost black', inherit=True)
s  = mpf.make_mpf_style(base_mpf_style='yahoo', marketcolors=mc)

# 設定頸線
seq_of_seq_of_points=[
    [(neckline_start_date,neckline_price),(neckline_end_date,neckline_price)],
    [(resistance_start_date,resistance_price),(resistance_end_date,resistance_price)]
                     ]

# 設定均線
added_plots={
    "SMA5": mpf.make_addplot(df_sma['SMA5'],width=0.4,color='xkcd:royal purple'),
    "SMA20": mpf.make_addplot(df_sma['SMA20'],width=0.4,color='xkcd:violet')
            }

# 繪製K線圖
kwargs = dict(type='candle', style=s, figratio=(19,10), volume=True, addplot=list(added_plots.values()), alines=dict(alines=seq_of_seq_of_points, linewidths=2.0, colors='xkcd:orange yellow', alpha=0.6), warn_too_much_data=1000)

mpf.plot(df_k_line,**kwargs)
```
