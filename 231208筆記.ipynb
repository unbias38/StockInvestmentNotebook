【我的解盤內容，適合那些投資人？】  
看我的分享內容，會因會投資類型不同，而擷取不同的內容，這是很正常的。至於甚麼人看我的解盤比較適合或是效果最大，當然是操作節奏和商品跟我一樣是最恰當，最好還能讀過我的書或是上過課，那加乘效果會更大。我的交易商品以個股為主，交易週期短則2-4周，長則1-2個月。因此，假設你交易商品和週期與我不同，那天天看我解盤，可能反而是有些危險的。其實只要是商品和週期與我不同，基本上都需要有能力自行轉接和調整，這裡只舉比較常見的兩種誤用類型。  
期貨跟選擇權：參考我的解盤作期貨或選擇權，因為我解盤都是用加權指數，其實本來期貨跟現貨就有誤差值。再者，解大盤是為了辨識多空、資金調節、風險控管，即使解盤誤差個50點、100點，對我也無傷大雅，但對作點數價差的投資人，影響就很大。第三，轉折時的效性，每天完整內容就一篇，周末影音要一周才一次，但倘若盤中有劇烈波動當下就要進行交易，等我的內容肯定是緩不濟急。其實參考我的解盤內容，做期權交易倒也不是行，確實也有一些同學，對照我解大盤的內容，操作期貨選擇權，效果也不錯，但自己要能轉換價差跟靈活操作。  
長期投資：承第一段，我的交易商品以個股為主，交易週期短則2-4周，長則1-2個月，而且資金和檔數會隨之調節。但長期投資的著重基本面、產業面，而且大多數時間都是滿持股，且沒有資金調節的習慣，一般都是從年初抱到年尾，但我的檔數和資金會進行波段調節，很多觀念跟策略甚至是衝突破的，甚至他根本不需要看每天的解盤。如此一來，我的解盤看法，由於先天條件的差異，可能會情緒波動忽上忽下，干擾持有部位的操作。  
所以不管你要看任何人的內容，都沒有關係，但請注意我說的大原則以及要留意的風險，選擇最適合你的解盤內容作參考，是相當重要的，否則即使解盤的再完整、再精細，都會因為彼此條件的不同，可能會適得其反。事實上我剛剛說的，算是投資的入門常識，但感覺很多新加入的同學不是很清楚，甚至看不懂我在寫甚麼，所以統一說明一下，希望我的分享能發揮最大的效果。
