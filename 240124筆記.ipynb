【不想抱股過年，可以怎麼處理】  
如果希望能夠專心吃年夜飯跟打麻將，不想還要一邊盯著美股或是富台指，決定不要抱股過年，那又該如何擬訂策略？首先，這要分成幾種類型，第一種是完全紀律而且意志堅定，那可以持續使用高資金水位，一路操作到農曆年前一周再積極減碼。  
第二種是現在仍然高持股，但不大紀律，交易技巧也不大好，那現在開始就應該漸進式的減碼，每周分批減一點，而且退出資金之後，維持現金狀態不再增加持股。如果中低持股者，可以保守經營到封關前一周。  
第三種，如果現在就已經空手，到封關前一周都找不到喜歡的個股，那就好好準備放假，因為即使你挑到好的個股，也要時間醞釀上漲，除非非常有把握，買完之後幾天內就會上漲，而且有一定漲幅，否則還是多看少做。  
為什麼一直強調前一周這件事情，因為封關前的賣壓，都是提早一周就出現，這段時間會容易小修正，然後封關前一兩天容易收紅。因此想要避開這些賣壓，最好封關一周之前就要先處理股票。
