【當行情不好時，該繼續選股交易，還是乾脆休息，或是反手做空？】  
其實這是個很複雜的問題，因為這牽涉到所謂「行情不好」的定義，究竟是短期空頭、中期空頭或是長期空頭，急跌、緩跌，還是盤整，以上各種情況都是所謂的行情不好，因為上漲的股票有限。  
如果把上述各種情況又加以排列組合，至少可以出現10多種以上的變化，而各種變化因應的策略各有不同，僅只能先以台股現況做分類，目前屬於短中期走空，急跌後的小盤整，長線多頭還沒破壞。  
由於現在還無法判斷是中期波段修正的「中繼站」還是「終點站」，所以最保守的投資人，暫時休息一下不是壞事，讓趨勢走的更明朗一些，再決定是否進場，畢竟從今年1月以來，只有資金高低調節，但沒有過多的天數是完全空手，就當股市放你的假，把握時間充分休息。  
但由於急跌暫告一段落，短線進入盤整，所以操作能力較強的投資人，還是有些個股可以選擇，但不只選股能力要強，還要有強大的心理素質，因為逆勢操作要有心理準備，失敗率可能會很高，必須搭配紀律停損的風險控管。  
最後則是積極投資人，選擇做頭或是20日線已經跌破下彎的個股，採取作空的策略，但因為短線小盤整，雖然趨勢屬於空方，但空方力道沒有上周強，所以最好還要搭配籌碼也是具備空方條件，這樣做空的勝率會比較高
