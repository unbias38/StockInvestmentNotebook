其實我說過很多次，指數只是決定做多做空策略、資金檔數調節、風險控管的工具，與個股交易是兩回事，指數有它的功能，但個股也有自己的趨勢、型態、位階、支撐、壓力、轉折。  
出現交易的訊號，就應該紀律執行，而不是等待指數上漲來拯救自己的股票。昨天指數急跌547點，但仍有個股抗跌，甚至逆勢上漲，同樣的即使今天指數止穩，還是有個股持續破底的。  
以我自己手上4檔個股，當然資金已經都壓到極低，加總起來只有約27萬，趨近於零持股，明細截圖昨天有分享在課程群組，但有兩檔是抗跌的，有兩檔偏弱，抗跌的就不用擔心，但轉弱的就要觀察是否出現停損條件，有碰觸到就要紀律停損。  
何況，我說過指數只是止穩，中期空方危機還沒解除，倘若20日線下彎，仍不排除只是弱勢橫盤幾天，之後有可能再度破底。
