【[櫃買指數與中小型股的關係](https://www.facebook.com/kevinyang20180801/posts/pfbid02uJ3SG2t9CgGUCMxWifRUFsdqpJVoji1BuwZyVEHbYLxsURBx72i3hSvfGr58SdqRl)】  
楊老師好，想請教有關櫃買指數的3個問題：1.如果手上的股票屬於上櫃股票，是不是可以只留意櫃買指數即可？2.查看加權指數與櫃買指數的走勢非常相近，是不是可以看成它們只有短期不同，中期與長期是相近？3.最近您提及中小型股要留意櫃買指數，這是包含上市的中小型股嗎？為何? 謝謝您。  
台灣的加權指數(又稱上市指數、集中市場)以及櫃買指數，也是一樣的邏輯。基本上加權指數就是表徵大型股的總體表現，而櫃買指數則是表徵中小型股的總體表現，這是無庸置疑的。大型股跟加權指數連動性較高，而中小型股與櫃買指數的連動性較高，因為兩種指數走勢不見得會完全同步，所以當你手中持有中小型股多時，受櫃買指數影響較大是沒有錯的。  
但投資人要知道，資金就像水，全世界乃至於台灣的資金，都是緊緊相連的，雖然櫃買指數與中小型股連動高，但不代表你可以完全不管加權指數，因為加權指數與櫃買指數兩者會互相牽引。而雖然單日漲跌、轉折點位，甚至短期波動，兩者確實會有落差，但是中長期走勢其實還是連動性高的，而且加權指數和櫃買指數輪動速度快，這一周加權指數強，下一周可能就變成櫃買指數強。  
區隔大型股與中小型股有3種方法，一種是掛牌的市場掛牌在加權指數定義為大型股，掛牌在櫃買指數定義為中小型股。第二種是股本，50億以上為大型股，50億以下為中小型股。最後一種則是市值，市值150億以上為大型股，150億以上為中小型股。但是在櫃買指數是1995年才開始，但是在此之前仍有股本小、市值低的中小型公司IPO，因此1955年之前掛牌的公司，即使中小型股，仍會掛牌在加權指數，但是這樣的個股，牽動他的漲跌，還是以櫃買指數為主。
